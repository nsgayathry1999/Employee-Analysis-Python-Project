## __4.Identify the predominant age group.__

```python
# First, make sure 'Age' is numeric
df_Copy['Age'] = pd.to_numeric(df_Copy['Age'], errors='coerce')

# Drop or fill NaNs if needed
df_Copy = df_Copy.dropna(subset=['Age'])  # or df_Copy['Age'].fillna(df_Copy['Age'].median(), inplace=True)

# Then apply pd.cut safely
bins = [20, 30, 40, 50, 60]
labels = ['21-30', '31-40', '41-50', '51-60']
df_Copy['age_group'] = pd.cut(df_Copy['Age'], bins=bins, labels=labels)

# Display result
age_group_distribution = df_Copy['age_group'].value_counts()
print("\nAge Group Distribution:\n", age_group_distribution)
print("Predominant Age Group:", age_group_distribution.idxmax())
```

```python
plt.figure(figsize=(6,4))
age_group_distribution.sort_index().plot(kind='bar', color='green')
plt.title("Age Group Distribution")
plt.xlabel("Age Group")
plt.ylabel("Number of Employees")
plt.tight_layout()
plt.show()
```

### Summary

#### _Most employees fall in the 31-40 range.

#### _This indicates a mid-career workforce, possibly with a good level of experience.

#### _Few employees are under 30, suggesting a potential gap in early-career hiring.

## __4.Highest salary expenditure by team and position.__

```python
team_salary = df.groupby('Team')['Salary'].sum().sort_values(ascending=False)
position_salary = df.groupby('Position')['Salary'].sum().sort_values(ascending=False)
print("\nTeam with Highest Salary Expenditure:\n", team_salary.head(1))
print("\nPosition with Highest Salary Expenditure:\n", position_salary.head(1))

```

```python
plt.figure(figsize=(8,6))
team_salary.plot(kind='bar', color='purple')
plt.title("Total Salary Expenditure by Team")
plt.xlabel("Team")
plt.ylabel("Total Salary")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,6))
position_salary.plot(kind='bar', color='brown')
plt.title("Total Salary Expenditure by Position")
plt.xlabel("Position")
plt.ylabel("Total Salary")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Summary

#### _The team with the highest salary expenditure is likely one of the core departments as C,PG

#### _Similarly, the position with the highest total salary might be Senior Managers or Department Heads.

#### _This concentration implies resource investment is skewed toward high-impact teams/roles.

## __5.Investigate if there's any correlation between age and salary, and represent it visually.__

```python
correlation = df['Age'].corr(df['Salary'])
print("\nCorrelation between Age and Salary:", round(correlation, 3))
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Age', y='Salary')
plt.title(f'Age vs Salary (Correlation: {round(correlation, 2)})')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Summary

#### _The correlation between age and salary is positive but not strong (e.g., ~0.2–0.4).

#### _This implies older employees tend to earn more, but age is not the only factor influencing salary.

#### _Other factors (like position, experience, or department) likely play a significant role.
